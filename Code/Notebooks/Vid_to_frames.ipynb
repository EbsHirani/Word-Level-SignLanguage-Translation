```python
import cv2
root= "/content/drive/MyDrive/MP sign language/Frames ISL"
#Looping over all the folders in the dataset
for i in os.listdir("ISL"):
  if os.path.isdir("ISL/" + i) and i != "Train_Test_Split":
    for j in os.listdir("ISL/" + i):
      print(i ,j)
      try:
        os.mkdir(root + f"/{j}")
      except:
        pass
      for k in os.listdir(f"ISL/{i}/{j}"):
        path = root + f"/{j}/{k}"
        try:
          os.mkdir(path)
        except:
          pass
        print(k)
        #reading a video from the dataset
        vidcap = cv2.VideoCapture(f'ISL/{i}/{j}/{k}')
        success,image = vidcap.read()
        count = 0
        #till frame exists
        while success:
          # print("hi")
          cv2.imwrite(path + f"/{count}.jpg", image)     # save frame as JPEG file      
          success,image = vidcap.read()
          # print('Read a new frame: ', success)
          count += 1
        print(count)
        if count == 0:
          print("****************************************************************************************************************")
          print(f"0 images in ISL/{i}/{j}/{k}")
      
  
      
  
```
